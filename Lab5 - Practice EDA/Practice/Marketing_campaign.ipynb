# Bài tập thực hành 1.3.3: Tìm hiểu và sử dụng SweetViz trên dữ liệu Marketing Campaign

# Import thư viện cần thiết

```python
import pandas as pd  # Để đọc và xử lý dữ liệu
import sweetviz as sv  # Thư viện SweetViz
```

# Bước 1: Đọc dữ liệu từ file CSV
# Giả sử file marketing_campaign.csv nằm cùng thư mục với notebook

```python
marketing_data = pd.read_csv("marketing_campaign.csv", sep="\t")  # Đọc dữ liệu từ file CSV
```

# Bước 2: Xử lý cơ bản dữ liệu (xử lý missing values nếu có)
# Ví dụ: Điền giá trị thiếu ở cột Income bằng median (để tránh bias)

```python
if marketing_data["Income"].isnull().sum() > 0:
    marketing_data["Income"] = marketing_data["Income"].fillna(marketing_data["Income"].median())

# Hiển thị info cơ bản để kiểm tra
print("Kích thước dữ liệu:", marketing_data.shape)
print("Thông tin cột:", marketing_data.info())
print("5 hàng đầu tiên:\n", marketing_data.head())
```

# Bước 3: Tạo báo cáo SweetViz
# Tạo report cho toàn bộ dataset (không so sánh)

```python
report = sv.analyze([marketing_data, "Marketing Campaign"])  # [dataset, tên report]
```

# Bước 4: Xuất báo cáo ra file HTML
# File sẽ được lưu trong thư mục hiện tại, mở bằng browser để xem

```python
report.show_html("sweetviz_marketing_report.html")
```

 Chú thích: 
 - sv.analyze() tự động phân tích tất cả cột (num, cat, date).
 - Nếu muốn so sánh (ví dụ: nhóm Response=1 vs Response=0):
   target = marketing_data[marketing_data["Response"] == 1]
   comparison = marketing_data[marketing_data["Response"] == 0]
   report = sv.compare([target, "Response=1"], [comparison, "Response=0"])
   report.show_html("sweetviz_comparison.html")

# Bài tập thực hành 1.3.4: Tìm hiểu và sử dụng AutoViz trên dữ liệu Marketing Campaign

# Import thư viện cần thiết

```python
import pandas as pd  # Để đọc và xử lý dữ liệu
from autoviz.AutoViz_Class import AutoViz_Class # Thư viện AutoViz
```

# Bước 1: Đọc dữ liệu từ file CSV

# Giả sử file marketing_campaign.csv nằm cùng thư mục

```python
marketing_data = pd.read_csv("marketing_campaign.csv", sep="\t")  # Đọc dữ liệu từ file CSV
```

# Bước 2: Xử lý cơ bản dữ liệu (tương tự SweetViz)

```python
if marketing_data["Income"].isnull().sum() > 0:
    marketing_data["Income"] = marketing_data["Income"].fillna(marketing_data["Income"].median())
```

# Hiển thị info cơ bản để kiểm tra

```python
print("Kích thước dữ liệu:", marketing_data.shape)
print("Thông tin cột:", marketing_data.info())
print("5 hàng đầu tiên:\n", marketing_data.head())
```

# Bước 3: Tạo biểu đồ AutoViz
# Sử dụng hàm AutoViz để tự động vẽ (verbose=1 để hiển thị tiến trình)

```python
viz = AutoViz_Class()
```

# Tham số: datapath (file path), max_cols_analyzed=10 (giới hạn cột để tránh overload), max_rows_analyzed=10000

```python
df = viz.AutoViz(
    filename="marketing_campaign.csv",
    sep="\t",                # ⚠ Nếu file gốc phân tách bằng tab thì để "\t", 
                             # nếu là CSV chuẩn thì để ",".
    depVar="", 
    dfte=None,
    header=0, 
    verbose=1, 
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=10000,
    max_cols_analyzed=10, 
    save_plot_dir=None
)
```

 Chú thích:
 - AutoViz đọc trực tiếp từ file CSV, không cần load vào DataFrame.
 - Nếu muốn dùng DataFrame: viz.AutoViz(None, dfte=marketing_data, max_cols_analyzed=10)
 - Kết quả: Hiển thị inline trong notebook (nhiều subplot).
 - Tùy chỉnh: Thêm chart_type='Line' để chỉ vẽ line chart, hoặc depVar='Income' để focus vào biến phụ thuộc.
