# 1.3/ Phân tích đơn biến và hai biến

# Bài toán 1: Phân tích đơn biến trên dữ liệu chim cánh cụt và giá nhà

# Nhiệm vụ 1: Sử dụng histogram, bar chart, pie chart để phân tích dữ liệu chim cánh cụt (culmen_length_mm, species)

# Import thư viện cần thiết

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Đọc dữ liệu từ file penguins_size.csv

```python
penguins_data = pd.read_csv("penguins_size.csv")
```

# 1. Vẽ histogram cho culmen_length_mm

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=penguins_data, x="culmen_length_mm", bins=20, kde=True)
plt.title("Phân bố chiều dài mỏ (culmen_length_mm) của chim cánh cụt")
plt.xlabel("Chiều dài mỏ (mm)")
plt.ylabel("Số lượng")
plt.show()
```

# 2. Vẽ bar chart cho species

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=penguins_data, x="species")
plt.title("Số lượng chim cánh cụt theo loài")
plt.xlabel("Loài")
plt.ylabel("Số lượng")
plt.show()
```

# 3. Vẽ pie chart cho species

```python
species_counts = penguins_data["species"].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Tỷ lệ phần trăm các loài chim cánh cụt")
plt.axis('equal')
plt.show()
```

# Nhiệm vụ 2: Sử dụng boxplot và violin plot để phân tích giá nhà (Price)

# Đọc dữ liệu từ file HousingPrices-Amsterdam-August-2021.csv

```python
housing_data = pd.read_csv("HousingPrices-Amsterdam-August-2021.csv")
```

# Xử lý giá trị thiếu (nếu có)

```python
housing_data = housing_data.dropna()
```

# 1. Vẽ boxplot cho Price

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=housing_data, y="Price")
plt.title("Boxplot của giá nhà tại Amsterdam")
plt.ylabel("Giá (EUR)")
plt.show()
```

# 2. Vẽ violin plot cho Price

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=housing_data, y="Price")
plt.title("Violin plot của giá nhà tại Amsterdam")
plt.ylabel("Giá (EUR)")
plt.show()
```

# Bài toán 2: Phân tích hai biến trên dữ liệu chim cánh cụt

# Nhiệm vụ 1: Vẽ scatter plot để phân tích mối quan hệ giữa culmen_length_mm và body_mass_g

# Vẽ scatter plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins_data, x="culmen_length_mm", y="body_mass_g", hue="species")
plt.title("Mối quan hệ giữa chiều dài mỏ và khối lượng cơ thể của chim cánh cụt")
plt.xlabel("Chiều dài mỏ (mm)")
plt.ylabel("Khối lượng (g)")
plt.legend(title="Loài")
plt.show()
```

# Nhiệm vụ 2-4: Sử dụng crosstab, pivot_table, và pairplot để khám phá mối quan hệ giữa species, sex, và các biến khác

# 2. Crosstab giữa species và sex

```python
crosstab_data = pd.crosstab(penguins_data["species"], penguins_data["sex"])
print("Crosstab giữa species và sex:")
print(crosstab_data)
```

# 3. Pivot table để tính trung bình culmen_length_mm theo species và sex

```python
pivot_data = pd.pivot_table(penguins_data, values="culmen_length_mm", index="species", columns="sex", aggfunc="mean")
print("\nPivot table trung bình culmen_length_mm:")
print(pivot_data)
```

# 4. Pairplot để khám phá mối quan hệ giữa các biến

```python
sns.pairplot(penguins_data, hue="species", vars=["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"])
plt.show()
```
